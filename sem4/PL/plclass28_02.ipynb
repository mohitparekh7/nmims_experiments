```python
import numpy as np 
import random
import timeit

n = 1000000
#return random floats in the half open interval [0.0,1.0]
x = [random.random() for i in range(n)]
y = [random.random() for i in range(n)]

z = [x[i]+y[i] for i in range(n)]
z[:3]
%timeit [x[i]+y[i] for i in range(n)]

xa = np.array(x)
ya = np.array(y)
za = xa + ya
za[:3]

%timeit xa + ya
```
