```python
from matplotlib import pyplot as plt
plt.style.use("classic")
slices=[25,30,45,90,21]
explode=[0.2,0.3,0,0.1,0.4]    #more than 5-6 data don;t use pie
labels=["JAVA",'PYTHON','CSS','HTML','C']
plt.pie(slices,labels=labels,explode=explode,shadow=True,startangle=90,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.title("my pie chart")   
```

```python
plt.style.use("fivethirtyeight")
minutes=[1,2,3,4,5,6,7,8,9]
player1=[1,2,3,4,2,3,3,5,4]
player2=[1,2,9,4,2,3,6,5,7]
labels=['player1','player2']
colors=["#6d904f",'#fc4f30']
plt.stackplot(minutes,player1,player2,labels=labels,colors=colors)
plt.legend(loc=(0.0,0.0))
plt.title("Mohit and pubg")
plt.tight_layout() #for perf image
```
