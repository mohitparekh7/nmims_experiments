```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
studentId = [1,2,3,4,5]
marks = [9,8,25,20,45]
adjacency = []
for i in range(len(marks)):
    x = []
    for j in range(len(marks)):
        value = abs(marks[i]-marks[j])
        x.append(value)
    adjacency.append(x)
print("Adjacency Matrix: ")
for i in range(len(adjacency)):
    print(adjacency[i])
```

```python
a = np.vstack((studentId, marks)).T
print(a)
```

```python
link = linkage(adjacency, 'single')
label = range(1,6)
plt.figure()
dendrogram(link, orientation = 'top', labels = label, distance_sort='descending', show_leaf_counts =True)
plt.show()
```
